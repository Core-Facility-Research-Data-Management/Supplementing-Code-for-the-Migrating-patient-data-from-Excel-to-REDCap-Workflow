### Purpose
This script is was used to generate synthetic data for the "Transform Data for REDCap Import"-workflow.<br>
It generates three CSV, one for each sheet in the final data XLSX.

```python
import random
import string
import datetime as dt
import csv

random.seed(42)

n_rows = 200

first_names = [
    "Anna","Ben","Clara","David","Eva","Felix","Greta","Hannah","Ivan","Julia",
    "Karl","Lena","Mia","Noah","Oskar","Paula","Quentin","Rita","Sophia",
    "Tom","Urs","Vera","William","Xenia","Yara","Zoe"
]

letters = list(string.ascii_uppercase)

def random_pseudonym(existing):
    while True:
        s = "".join(random.choices(string.ascii_letters + string.digits, k=10))
        if s not in existing:
            existing.add(s)
            return s

def random_date(start_year, end_year):
    start = dt.date(start_year, 1, 1)
    end = dt.date(end_year, 12, 31)
    delta_days = (end - start).days
    d = start + dt.timedelta(days=random.randint(0, delta_days))
    return d

def format_date(d):
    return d.strftime("%d.%m.%Y")

rows = []
seen_pseudos = set()

for _ in range(n_rows):
    pseudo = random_pseudonym(seen_pseudos)

    first = random.choice(first_names)
    surname = random.choice(letters) + "."

    dob = random_date(1900, 2020)
    incl = random_date(2022, 2025)

    # Age at inclusion in years, one decimal, comma as decimal separator
    age_years = (incl - dob).days / 365.25
    age_str = f"{age_years:.1f}".replace(".", ",")

    sex = random.choice(["1", "2", "male", "female"])

    rows.append([
        pseudo,
        first,
        surname,
        format_date(dob),
        sex,
        format_date(incl),
        age_str
    ])

header = [
    "Pseudonym",
    "First name",
    "Surname",
    "Date of birth",
    "Sex",
    "Inclusion date",
    "Age at inclusion"
]

with open("synthetic_patients.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="|")
    writer.writerow(header)
    writer.writerows(rows)
```

```python
import random
import csv
import datetime as dt

random.seed(2025)

n_rows = 200

therapy_values = [0, 1, 2]
drugs_values = [1, 2, 3, 4, "ASA", "Paracetamol", "Ibuprofen", "Morphine"]
surgery_values = [1, 2, 3]

phrases_subject = [
    "Patient", "The patient", "Subject", "This individual", "The participant"
]
phrases_status = [
    "shows stable vital signs",
    "reports mild chest discomfort",
    "has improved exercise tolerance",
    "complains of intermittent headaches",
    "denies shortness of breath",
    "presents with elevated blood pressure",
    "has well-controlled diabetes",
    "shows no signs of acute infection",
    "reports improved sleep quality",
    "has persistent fatigue"
]
phrases_plan = [
    "continue current therapy",
    "adjust medication dosage",
    "schedule follow-up in two weeks",
    "monitor blood pressure daily",
    "consider cardiology consultation",
    "encourage regular physical activity",
    "review medication adherence",
    "repeat laboratory tests next month",
    "evaluate response to treatment",
    "provide additional patient education"
]

def random_date(start_year, end_year):
    start = dt.date(start_year, 1, 1)
    end = dt.date(end_year, 12, 31)
    delta_days = (end - start).days
    return start + dt.timedelta(days=random.randint(0, delta_days))

def format_date(d):
    return d.strftime("%d.%m.%Y")

def random_sentence():
    s = random.choice(phrases_subject)
    st = random.choice(phrases_status)
    pl = random.choice(phrases_plan)
    return f"{s} {st}; {pl}."

rows = []

for _ in range(n_rows):
    therapy = random.choice(therapy_values)
    drug = random.choice(drugs_values)
    surgery = random.choice(surgery_values)

    start = random_date(2026, 2027)

    # Ensure end is at least 10 days after start, still within 2027
    min_end = start + dt.timedelta(days=10)
    max_end = dt.date(2027, 12, 31)
    if min_end > max_end:
        start = max_end - dt.timedelta(days=10)
        min_end = start + dt.timedelta(days=10)
    delta_days = (max_end - min_end).days
    end = min_end + dt.timedelta(days=random.randint(0, max(0, delta_days)))

    note = random_sentence()

    rows.append([
        therapy,
        drug,
        surgery,
        format_date(start),
        format_date(end),
        note
    ])

header = [
    "Therapy",
    "Drugs",
    "Surgery",
    "Therapy start",
    "Therapy end",
    "Therapy note"
]

with open("synthetic_therapy.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="|")
    writer.writerow(header)
    writer.writerows(rows)
```
